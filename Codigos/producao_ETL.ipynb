```python
#CRIANDO UM DATAFRAME A PARTIR DO ARQUIVO CSV||
df_prod = pd.read_csv('/content/drive/MyDrive/Tech Challenge 1 modulo - FIAP/Dados Brutos/Producao.csv',sep=',',header=None)
```

```python
df_prod.head()
```

```python
#RENOMEANDO AS COLUNAS
ano = 1970
df_prod.rename(
    columns={0: 'id', 1: 'categoria', 2: 'tipo'}
    | {col: f'quantidade(l)_{ano + (col - 3)}' for col in df_prod.columns[3:]},
    inplace=True
)
```

```python
df_prod.head()
```

```python
#PADRONIZNADO A NOMENCLATURA DAS LINHAS DAS COLUNAS CATEGORIA E TIPO
df_prod[['categoria','tipo']] = df_prod[['categoria','tipo']].applymap(lambda x: x.lower())
```

```python
#ANALISANDO SE EXISTEM VALORES DUPLICADOS PARA CATEGORIAS
df_prod['categoria'].value_counts()
```

```python
#DECLARANDO LISTA COM OS TIPOS DE VINHOS
tipos_vinhos = ['tinto','branco','rosado','vinho leve','vinho licoroso','vinho composto','vinho orgânico']
```

```python
#CRIANDO NOVA VARIAVEL PARA ARMAZENAR SOMENTE OS TIPOS DE VINHOS DESEJADOS
df_prod_vinhos = df_prod[df_prod['tipo'].isin(tipos_vinhos)]
```

```python
#ANALISANDO ESTATISTICA DESCRITIVA
df_prod_vinhos.describe()
```

```python
df_prod_vinhos['categoria']
```

```python
#TRATANDO VALORES DA COLUNA DE CATEGORIA
mapeando = {0:'vinho de mesa','vm_tinto':'vinho de mesa','vm_branco':'vinho de mesa','vm_rosado':'vinho de mesa',
            'vinho fino de mesa (vinífera)':'vinho fino mesa','vv_tinto':'vinho fino mesa','vv_branco':'vinho fino mesa','vv_rosado':'vinho fino mesa'
            ,'de_vinho leve':'derivados','de_vinho licoroso':'derivados','de_vinho composto':'derivados','de_vinho orgânico':'derivados'}
df_prod_vinhos['categoria'] = df_prod_vinhos['categoria'].map(mapeando).fillna(df_prod_vinhos['categoria'])
```

```python
#TRATANDO VALORES DA COLUNA DE id
mapeando_id = {'vinho de mesa':0,'vinho fino mesa':1,'derivados':2}
df_prod_vinhos['id'] = df_prod_vinhos['categoria'].map(mapeando_id).fillna(df_prod_vinhos['id'])
```

```python
#MUDANDO O dtype da coluna de id
df_prod_vinhos['id'] = df_prod_vinhos['id'].astype(int)
```
