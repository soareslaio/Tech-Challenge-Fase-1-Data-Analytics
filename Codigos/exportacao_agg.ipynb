```python
#IMPORTANDO BIBLIOTECAS NECESSARIAS
import pandas as pd
```

```python
#LENDO ARQUIVO .CSV E CRIANDO DataFrame
export_df = pd.read_csv('/content/drive/MyDrive/Tech Challenge 1 modulo - FIAP/Dados Tratados/expt.csv',sep=',')
```

```python
#ANALISANDO O DataFrame CRIADO ANTERIORMENTE
export_df.head()
```

```python
#AJUSTANDO DataFrame PARA FACILITAR A MANIPULAÇÃO NO POWERBI
colunas_quantidade = [f'quantidade_{i}' for i in range(2008, 2023)]
colunas_valor = [f'valor_{i}' for i in range(2008, 2023)]
df_agregado = export_df.groupby('Paises')[colunas_quantidade + colunas_valor].sum().reset_index()
df_agregado['quantidade_total'] = df_agregado[colunas_quantidade].sum(axis=1)
df_agregado['valor_total'] = df_agregado[colunas_valor].sum(axis=1)
df_agregado = df_agregado[['Paises', 'quantidade_total', 'valor_total']]
print(df_agregado)
```

```python
#AGRUPANDO COLUNAS DE FORMA DESCENDENTE DE ACORDO COM A COLUNA valor_total
df_vinhosagregado = df_agregado.sort_values(by='valor_total',axis=0,ascending=False)
```

```python
#EXPORTANTO DataFrame AGREGADO
df_vinhosagregado.to_csv('export_vinhosagg.csv', index=False)
```
