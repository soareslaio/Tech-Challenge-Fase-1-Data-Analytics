```python
#LENDO DADOS DO ARQUIVO CSV DE EXPORTAÇÂO
df = pd.read_csv("/content/drive/MyDrive/Tech Challenge 1 modulo - FIAP/Dados Brutos/imp.csv",sep=";", header=None)
df.head(2)
#CRIANDO NOVO CABECALHO
df.drop(df.columns[0], axis=1, inplace=True)
#RENOMENADO NOME DAS COLUNAS A PARTIR DA 2 COLUNA USANDO LOOP
for i in range(2, df.shape[1] + 2, 2):
    ano = 1970 + ((i - 2) // 2)
    df = df.rename(columns={i: f'quantidade_{ano}', i + 1: f'valor_{ano}'})
#RENOMEIA A PRIMEIRA COLUNA
df = df.rename(columns={df.columns[0]: 'Paises'})
#EXIBINDO DATAFRAME RESULTANTE
df = df.drop(df.index[0])
df.tail(6)
```

```python
#ANALISANDO DIMENSOES DO DataFrame
df.shape
```

```python
#CRIANDO DOIS DF: df_valor SOMENTE COM OS PAISES E AS COLUNAS COM "valor" E df_quantidade SOMENTE COM OS PAISES E AS COLUNAS COM "quantidade"
colunas_selecionadas = ['Paises'] + [coluna for coluna in df.columns if coluna.startswith('valor_')]
df_valor = df[colunas_selecionadas]
colunas_selecionadas1 = ['Paises'] + [coluna for coluna in df.columns if coluna.startswith('quantidade_')]
df_quantidade = df[colunas_selecionadas1]
```

```python
#VERIFICANDO NULOS DO DataFrame df_quantidade E PREENCHENDO COM 0
df_quantidade = df_quantidade.fillna(0)
df_quantidade.info()
```

```python
#VERIFICANDO NULOS DO DataFrame df_valor E PREENCHENDO COM 0
df_valor = df_valor.fillna(0)
df_valor.info()
```

```python
#VALIDANDO VALORES NULOS
print(df_valor.isnull().sum().sum())
print(df_quantidade.isnull().sum().sum())
```

```python
#EXPORTANTO ARQUIVO .CSV
df_valor.to_csv('imp_valor.csv', index=False)
df_quantidade.to_csv('imp_quantidade.csv', index=False)
```

```python
#UTILIZANDO MELT PARA TRANSFORMAR O DataFrame EM FORMATO LONGO
df_quantidade_long = pd.melt(df_quantidade, id_vars=['Paises'], var_name='Ano', value_name='Quantidade')

#ENCONTRANDO AS LINHAS COM OS VALORES MÁXIMOS EM CADA GRUPO 'Paises''
df_quantidade_max = df_quantidade_long.loc[df_quantidade_long.groupby('Paises')['Quantidade'].idxmax()]

#EXIBINDO DataFrame RESULTANTE
df_quantidade_max
```

```python
#SOMANDO TODOS OS VALORES PARA CADA LINHA E CRIANDO UMA COLUNA CHAMADA "Total" COM O SOMATORIO DA QUANTIDADE EXPORTADA PARA CADA PAÍS
df_quantidade['Total'] = df_quantidade.iloc[:, 1:].sum(axis=1)

#EXIBINDO O DataFrame RESULTANTE E ORDENANDO DE FORMA DECRESCENTE
df_quantidade = df_quantidade.sort_values(by='Total', ascending=False)
```
